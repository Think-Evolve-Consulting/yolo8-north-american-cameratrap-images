## Check the Model Size

```python
!ls -l {HOME}/runs/detect/train/weights/best.pt
print("\nFile Size:", 6246873 / 1000000, "MB")
```

```python
import torch
import sys

# Load the model
model = torch.load('runs/detect/train/weights/best.pt', map_location=torch.device('cpu'))

# Get the size of the model in bytes
model_size_bytes = sys.getsizeof(model)

# Convert bytes to megabytes
model_size_megabytes = model_size_bytes /  (1024*1024)

print(f"Size of the model: {model_size_megabytes} MB")
```
